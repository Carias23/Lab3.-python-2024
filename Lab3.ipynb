```python
# 5) Agregar nuevas coordenadas al array y mostrar las dimensiones
coord_array = np.array([[1.33, 4.5],
                        [30.0, 10.7],
                        [70.2, 0.5]])

new_coords = np.array([[37.1, -3.6]])

# Agregar nuevas coordenadas al array
updated_array = np.vstack([coord_array, new_coords])

# Mostrar las dimensiones del nuevo array
print("Dimensiones del nuevo array:", updated_array.shape)

```

```python
# 6) Copiar el array y transponerlo, luego agregar nuevas coordenadas
copied_array = np.copy(updated_array)

# Transponer el array
transposed_array = np.transpose(copied_array)

# Agregar nuevos pares de coordenadas
new_coords_2 = np.array([[10.8, 3.0],
                         [35.8, 12.0]])

# Agregar nuevas coordenadas al array transpuesto
transposed_array = np.hstack([transposed_array, new_coords_2])

# Transponer nuevamente para obtener el formato original
final_array = np.transpose(transposed_array)

print("Nuevo array tras las operaciones:", final_array)
```

```python
# 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid
# Primero, calculamos la distancia en grados entre cada municipio y Madrid (no consideramos Madrid en el cálculo)
def calcular_distancia(lat1, lon1, lat2, lon2):
    # Convertir grados a radianes
    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)
    # Fórmula haversine
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    # Radio de la Tierra en kilómetros
    r = 6371
    return c * r

# Coordenadas de Madrid
madrid_lat = 40.4168
madrid_lon = -3.7038

# Calcular la distancia de cada municipio a Madrid
df['Distancia_a_Madrid'] = calcular_distancia(df['Y'], df['X'], madrid_lat, madrid_lon)

# Encontrar el municipio más cercano y más lejano a Madrid
municipio_mas_cercano_madrid = df[df['Distancia_a_Madrid'] == df['Distancia_a_Madrid'].min()]['Municipio'].values[0]
municipio_mas_lejano_madrid = df[df['Distancia_a_Madrid'] == df['Distancia_a_Madrid'].max()]['Municipio'].values[0]

print("8) Municipio más cercano a Madrid:", municipio_mas_cercano_madrid)
print("   Municipio más lejano a Madrid:", municipio_mas_lejano_madrid)
```
